```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
```

```python
image = mpimg.imread("./images/horse.jpg")
print("image shape: ", image.shape)

plt.imshow(image)
img_plot = plt.imshow(image)
plt.axis('off')
```

```python
image_flattened = image.reshape((image.shape[0] * image.shape[1], image.shape[2]))
print(image_flattened.shape)
```

```python
num_clusters = 10
kmeans_model = KMeans(n_clusters=num_clusters).fit(image_flattened)
print(kmeans_model)

labels = kmeans_model.predict(image_flattened)
compressed_image = np.zeros_like(image_flattened)
for k in range(num_clusters):
    compressed_image[labels == k] = kmeans_model.cluster_centers_[k]

compressed_image = compressed_image.reshape(image.shape)
```

```python
img_plot = plt.imshow(compressed_image)
plt.axis('off')
```
