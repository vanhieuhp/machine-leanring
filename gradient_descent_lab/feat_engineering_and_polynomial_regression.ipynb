```python
import numpy as np
import matplotlib.pyplot as plt
from additional_utils.lab_utils_multi import zscore_normalize_features, run_gradient_descent_feng

np.set_printoptions(precision=2)
```

```python
# create target data
x = np.arange(0, 20, 1)
y = 1 + x ** 2
X = x.reshape(-1, 1)

model_w, model_b = run_gradient_descent_feng(X, y, iterations=1000, alpha=1e-2)
plt.scatter(x, y, marker='x', c='r', label='Actual Value')
plt.title("No feature engineering")
plt.plot(x, X @ model_w + model_b, label="Predicted Value")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
# create target data
x = np.arange(0, 20, 1)
y = 1 + x ** 2

# engineer features
X = x ** 2  #  <--- added engineered feature
```

```python
X = X.reshape(-1, 1)
model_w, model_b = run_gradient_descent_feng(X, y, iterations=1000, alpha=1e-5)
plt.scatter(x, y, marker='x', label="Actual Value")
plt.title("Added x**2 feature")
plt.plot(x, np.dot(X, model_w) + model_b, label="Predicted Value")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
x = np.arange(0, 20, 1)
y = x ** 2

# engineer features
X = np.c_[x, x ** 2, x ** 3]
X_features = ['x', 'x^2', 'x^3']
```

```python
fig, ax = plt.subplots(1, 3, figsize=(12, 3), sharey=True)
for i in range(len(ax)):
    ax[i].scatter(X[:, i], y)
    ax[i].set_xlabel(X_features[i])
ax[0].set_ylabel("y")
plt.show()
```

```python
x = np.arange(0, 20, 1)
X = np.c_[x, x ** 2, x ** 3]
print(f"Peak to Peak range by column in Raw              X:{np.ptp(X, axis=0)}")

# add mean_normalization
X = zscore_normalize_features(X)
print(f"Peak to Peak range by colun in Normalized        X:{np.ptp(X, axis=0)}")
```

```python
x = np.arange(0, 20, 1)
y = x ** 2

X = np.c_[x, x ** 2, x ** 3]
X = zscore_normalize_features(X)

model_w, model_b = run_gradient_descent_feng(X, y, iterations=100000, alpha=1e-1)
plt.scatter(x, y, marker='x', c='r', label="Actual Value")
plt.title("Normalized x x**2, x**3 feature")
plt.plot(x, X @ model_w + model_b, label="Predicted Value")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
